## Define the following terms

    A) data
        * A raw facts
            * These facts can range from numbers, names, dates, etc.
    B) field
        * A character or group of characters with a specific meaning
            * This defines and stores data
    C) record
        * Logically connected set of one more fields that describes something
            * For instance, a record can contain information regarding someone's age, name, occupation, income, credit score, etc. (information pertaining to the person)
    D) file
        * Collection of related records
            * This could be records about workers that share the same occupation such as professionals within IT or more specifically data science.
